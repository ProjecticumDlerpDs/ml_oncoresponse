## merge()
In deze stap leer ik hoe ik de `merge()`-functie in pandas werkt. Een **inner join** combineert twee datasets op een kolom of index, en geeft alleen de rijen terug die in beide datasets voorkomen. Ik maak eerst een nieuwe DataFrame met alleen de neerslaggegevens van één weerstation. Dit helpt om een kleiner en meer gerichte dataset te krijgen.

```python
precip_one_station = climate_precip.query("STATION == 'GHCND:USC00045721'")
precip_one_station.head()
```

```python
precip_one_station.shape
```

Nu ga ik met behulp van `pd.merge()` de neerslagdata van één station ('precip_one_station`) combineren met de tepmeratuurdata (`climate_temp`). Pandas kijkt automatisch naar kolommen die in beide Dataframes voorkomen (hier `STATION`) en gebruikt deze als sleutel voor de merge. Het resulataat is een nieuwe tabel (`inner_merged`) waarin alleen de rijen staan voor het geselecteerde station, gecombineerd met de bijbehorende temperatuurgegevens. 

```python
inner_merged = pd.merge(precip_one_station, climate_temp)
inner_merged.head()
```

Met `inner_merged.head()` kan ik de eerste vijf rijen bekijken om te controleren of de merge goed is uitgevoerd.

```python
inner_merged.shape
```

Stel je wilt niet alleen één station combineren, maar beide volledige datasets samenvoegen, en je wilt precies aangeven op welke kolommen de merge moeten gebeuren. dan kun je de parameter `on` gebruiken en de kolommen opgeven waarop je wilt joinen.
Op die manier kun je bijvoorbeeld zowel `STATION` als `DATE` als sleutels gebruiken om de datasets correct te maken.


```python
inner_merged_total = pd.merge(
    climate_temp, climate_precip, on=["STATION", "DATE"]
)
inner_merged_total.head()
```

```python
inner_merged_total.shape
```

Na deze merge zie ik dat de nieuwe DataFrame 48 kolommen bevat. Dit komt doordat ik met `on=["STATION","DATE"]` expliciet heb aangegeven welke kolommen met dezelfde naam in beide datasets apart bestaan. Die kolommen krijgen automatisch de achtervoegsels `_x` en `_y` waardoor er uiteindelijk één kolom meer overblijft. 

```python
## Outer join
Met een outer joint blijven ook rijen zonder match behouden. In dit voorbeeld gebruik ik de kleinere neerslag-DataFrame `precip_one_station` en de volledige `climate_temp` DataFrame, en ik join op de kolimmen `STATION` en `DATE`.
```

```python
outer_merged = pd.merge(
    precip_one_station, climate_temp, how="outer", on=["STATION", "DATE"]
)
outer_merged.head()
```
