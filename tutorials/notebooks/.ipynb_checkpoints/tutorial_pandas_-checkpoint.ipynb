Dit notebook laat zien dat ik de RealPython tutorial over het combineren van data in pandas heb doorgewerkt. 


Ik begin met het importeren van de pandas-library zodat ik pd.read_csv(),pd.merge(), etc kan gebruiken. Vervolgens stel ik in hoeveel kolommen pandas laat zien als ik een DataFrame print (50).

```python
import pandas as pd

pd.set_option("display.max_columns", 50)
```

### Data importeren

Hier laad ik de dataset met temperatuur- en neerslaggegevens in DataFrames. Met `.head()` bekijk ik de eerste 5 rijen van de DataFrame, zodat ik een idee krijg van de structuur van de data. Met `shape` kan ik het aantal tijden en kolommen zien. Dit is handig om te controleren voordat ik verder ga met het analyseren van de data.



```python
climate_temp = pd.read_csv("tutorial_data/climate_temp.csv")
climate_temp.head()
```

```python
climate_temp.shape
```

```python
climate_precip = pd.read_csv("tutorial_data/climate_precip.csv")
climate_precip.head()
```

```python
climate_precip.shape
```

```python
precip_one_station = climate_precip.query("STATION == 'GHCND:USC00045721'")
precip_one_station.head()
```

```python
precip_one_station.shape
```

Nu ga ik met behulp van `pd.merge()` de neerslagdata van één station ('precip_one_station`) combineren met de tepmeratuurdata (`climate_temp`). Pandas kijkt automatisch naar kolommen die in beide Dataframes voorkomen (hier `STATION`) en gebruikt deze als sleutel voor de merge. Het resulataat is een nieuwe tabel (`inner_merged`) waarin alleen de rijen staan voor het geselecteerde station, gecombineerd met de bijbehorende temperatuurgegevens. 

```python
inner_merged = pd.merge(precip_one_station, climate_temp)
inner_merged.head()
```

Met `inner_merged.head()` kan ik de eerste vijf rijen bekijken om te controleren of de merge goed is uitgevoerd.

```python
inner_merged.shape
```

Stel je wilt niet alleen één station combineren, maar beide volledige datasets samenvoegen, en je wilt precies aangeven op welke kolommen de merge moeten gebeuren. dan kun je de parameter `on` gebruiken en de kolommen opgeven waarop je wilt joinen.
Op die manier kun je bijvoorbeeld zowel `STATION` als `DATE` als sleutels gebruiken om de datasets correct te maken.


```python
inner_merged_total = pd.merge(
    climate_temp, climate_precip, on=["STATION", "DATE"]
)
inner_merged_total.head()
```

```python
inner_merged_total.shape
```

Na deze merge zie ik dat de nieuwe DatFrame 48 kolommen bevat. Dit komt doordat ik met `on=["STATION","DATE"]` expliciet heb aangegeven welke kolommen met dezelfde naam in beide datasets apart bestaan. Die kolommen krijgen automatisch de achtervoegsels `_x` en `_y` waardoor er uiteindelijk één kolom meer overblijft.
